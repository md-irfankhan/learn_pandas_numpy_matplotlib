```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tipscorr=tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

```python
sns.clustermap(tipscorr.corr(),annot=True)
```

```python
pvtflight=flights.pivot_table(index='month',columns='year',values='passengers')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
