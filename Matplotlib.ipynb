```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
x
y=x**3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(11,10),dpi=100)
axes1=fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y)
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y)
```
